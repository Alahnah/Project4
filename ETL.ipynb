```python
import pandas as pd
import numpy as np
```

```python
#Read the data into a Pandas DataFrame
job_postings_df = pd.read_csv('fake_job_postings.csv')
job_postings_df.head()
```

```python
#Get the information
job_postings_df.info()
```

```python
##Create Country DataFrame
```

```python
# Split the location column into country and subregion columns
job_postings_df[['country', 'subregion']] = job_postings_df['location'].str.split(',', n=1, expand=True)

job_postings_df.drop(columns=['location'], inplace= True)
job_postings_df.head()
```

```python
# Split the subregion column into state/province and city columns
job_postings_df[['region', 'city']] = job_postings_df['subregion'].str.split(',', n=1, expand=True)

job_postings_df.drop(columns=['subregion'], inplace= True)
job_postings_df.head()
```

```python
#Export as new CSV
job_postings_df.to_csv('job_postings_transformed.csv', index=False)
```

```python
#Get unique countries, regions, and cities in separate lists
countries = job_postings_df['country'].unique().tolist()
regions = job_postings_df['region'].unique().tolist()

print(countries)
print(regions)
```

```python
#Get number of distinct values in countries and regions
print(len(countries))
print(len(regions))
```

```python
##Transform original dataset for use in SQL database
```

```python
job_postings_transformed_df = pd.read_csv('job_postings_transformed.csv')
job_postings_transformed_df.head()
```

```python
#Print info
job_postings_transformed_df.info()
```

```python
# Convert integer columns to boolean
columns_to_convert = ['telecommuting', 'has_company_logo', 'has_questions', 'fraudulent']
job_postings_transformed_df[columns_to_convert] = job_postings_transformed_df[columns_to_convert].astype(bool)
job_postings_transformed_df.info()
```

```python
# Merge the country_df with the job_postings_transformed_df on the "country" column
job_merged_df = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge
                         (job_postings_transformed_df, country_df, on="country", how="left"),
                         employment_type_df, on="employment_type", how="left"),
                         required_experience_df, on="required_experience", how="left"),
                         required_education_df, on="required_education", how="left"),
                         industry_df, on="industry", how="left"),
                         function_df, on="function", how="left")
                        
# Display the resulting merged DataFrame
job_merged_df.head(25)
```

```python
#Drop unnecessary columns

columns_to_drop = ['country', 'region', 'city', 'company_profile', 'description', 'requirements', 'benefits', 'employment_type', 'required_experience', 'required_education', 'industry','function']

# Use the drop method to remove these columns from the DataFrame
job_merged_transformed_df = job_merged_df.drop(columns=columns_to_drop, axis=1)
job_merged_transformed_df.head()
```

```python
#Export as CSV
job_merged_transformed_df.to_csv('job_merged_transformed_df.csv', index=False)
```
